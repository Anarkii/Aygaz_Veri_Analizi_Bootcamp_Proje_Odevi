```python
sns.barplot(x="neighbourhood_group", y=df["neighbourhood_group"].index,data=df, color = 'green').set_title("3")
```

```python
sns.barplot(x="neighbourhood_group",
            y="price",
            hue="room_type",
            data=df)
plt.title("4")
```

```python
df.groupby('neighbourhood_group').sum().plot.pie(y='price', autopct="%.1f%%", ylabel="", legend=False, figsize=(10,10))
```

```python
cols = ["x", "y", "z"]
for col in cols:
  plt.figure(figsize=(10, 10))
  sns.scatterplot(x="minimum_nights", y="price", data=df)
  plt.show()
```

```python
sns.histplot(df["neighbourhood_group"], kde=True)
```

```python
sns.jointplot(x="price", y=df["minimum_nights"], data=df)
```

```python
sns.catplot(x="neighbourhood_group", y="price", data=df)
plt.title("5")
```
